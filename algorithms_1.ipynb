```python
import numpy as np
import matplotlib.pyplot as plt
import timeit
import math
```

```python
def time_execute(function, n):
  timings = []
  for i in range(n):
    t = 0
    for j in range(5):
      v = np.random.random(i)
      start = timeit.default_timer()
      function(v)
      end = timeit.default_timer()
      t += end - start
    timings.append(t / 5)
  return timings
```

```python
def constant(v):
  return 1
def sum_of_elements(v):
  sum = 0
  for i in v:
    sum += i
  return sum
def prod_of_elements(v):
  prod = 0
  for i in v:
    prod *= i
  return prod
def polynom(v):
  x = 1.5
  y = 0
  for k in range(len(v)):
      y += v[k] * x ** k
  return y
def polynom_horner(v):
  x = 1.5
  y = 0
  for k in range(len(v), 0, -1):
      y = v[k-1] * x * y
  return y
def bubblesort(v):
  for i in range(len(v)-1):
      for j in range(len(v)-1):
          if(v[j] > v[j+1]):
              v[j], v[j+1] = v[j+1], v[j]
  return v
def quicksort(v):
  if len(v) <= 1:
      return v
  else:
      pivot = v[0]
      left = [x for x in v[1:] if x < pivot]
      right = [x for x in v[1:] if x >= pivot]
  return quicksort(left) + [pivot] + quicksort(right)
def timsort(v):
  return np.sort(v, kind='stable')
```

```python
N = 2000
```

```python
t_1 = time_execute(constant, N)
```

```python
t_2 = time_execute(sum_of_elements, N)
```

```python
t_3 = time_execute(prod_of_elements, N)
```

```python
t_4 = time_execute(polynom, 1500)
```

```python
t_5 = time_execute(polynom_horner, N)
```

```python
t_6 = time_execute(bubblesort, 500)
```

```python
t_7 = time_execute(quicksort, N)
```

```python
t_8 = time_execute(timsort, N)
```

```python
n = list(range(N))
k = np.median(t_1)

plt.plot(n, t_1)
plt.plot([0, N], [k, k])
plt.xlabel('N')
plt.ylabel('Time')
plt.legend(['Constant', 'O(1)'])
plt.show()
```

```python
def time_execute_matrix(n):
  timings = []
  for i in range(n):
    t = 0
    for j in range(5):
      A = np.random.random((i,i))
      B = np.random.random((i,i))
      start = timeit.default_timer()
      A @ B
      end = timeit.default_timer()
      t += end - start
    timings.append(t / 5)
  return timings
```

```python
t_9 = time_execute_matrix(500)
```

```python
n = list(range(500))
k = np.median([t_9[i] / ((i+1) ** 3) for i in n])

plt.figure(figsize=(10,5))
plt.plot(n, t_9)
plt.plot(n, [k * ((i+1) ** 3) for i in n])
plt.xlabel('N')
plt.ylabel('Time')
plt.legend(['Matrix product', 'O(n^3)'])
plt.show()
```
